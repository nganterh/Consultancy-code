```python
import time
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait

investing_url = 'https://www.investing.com'

for row in range(df_main.shape[0]):
    browser.get(f'{investing_url}{df_main.at[row, "url"]}-historical-data')
    time.sleep(0.5)

    WebDriverWait(browser, 5).until(EC.presence_of_element_located((By.ID, 'widgetFieldDateRange'))).click()
    date_input = browser.find_element_by_id('startDate')

    WebDriverWait(browser, 5).until(EC.presence_of_element_located((By.ID, 'startDate'))).clear()
    date_input.send_keys('01/01/1990')

    WebDriverWait(browser, 5).until(EC.element_to_be_clickable((By.CSS_SELECTOR, '#applyBtn'))).click()
    time.sleep(3)
    
    df_list = pd.read_html(browser.page_source, attrs={'class':'genTbl closedTbl historicalTbl',
                                                       'id':'curr_table'}, keep_default_na=False)
    df_stock = pd.concat(df_list)
    break
    
display(df_stock)
```
